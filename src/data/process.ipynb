```python
# Square center cropping

def squareCenterCrop(img):
    """
    Perform a center cropping of an image.
    Args:
        img_path
    
    """
    height, width, _ = img.shape
    size = min(height, width)
    left = (width - size) // 2
    top = (height - size) // 2
    right = left + size
    bottom = top + size
    img_cropped = img[top:bottom, left:right]



```
